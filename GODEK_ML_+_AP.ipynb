```python
"""
Participants in the older group showed substantially lower performance in LPS,
reflecting age-related decline in fluid intelligence.
"""
```

```python
mean_lps = df['LPS score'].mean()
std_lps  = df['LPS score'].std()
```

```python
df['LPS z-scale'] = (df['LPS score'] - mean_lps) / std_lps
```

```python
def Zscore(Xscore):
  mean = df[Xscore].mean()
  std = df[Xscore].std()
  return (df[Xscore] - mean) / std


# here I created function so i can do same code twice with proper call.
```

```python
df['LPS z-scale2'] = Zscore('LPS score')
```

```python
df['TMT A rev'] = 5 - df['TMT A scale']
df['TMT B rev'] = 5 - df['TMT B scale']

# Now 4 = best, 1 = worst
```

```python
# old way:
df['TMT A z-scale'] = (df['TMT A rev'] - df['TMT A rev'].mean()) / df['TMT A rev'].std()
df['TMT B z-scale'] = (df['TMT B rev'] - df['TMT B rev'].mean()) / df['TMT B rev'].std()

# now with function usage:
df['TMT A z-scale2'] = Zscore('TMT A rev')
df['TMT B z-scale2'] = Zscore('TMT B rev')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix

# those imports are not at the beginning to show where machine learning comes in. Those imports are used for 1) Testing, 2) Logistic Regression, 3) resaults
```

```python
# Features
X = df[['LPS z-scale2', 'WST z-scale', 'TMT A z-scale2', 'TMT B z-scale2']]

# Target
y = df['Age group'].map({'20-35':0, '55-80':1}) # 0 = young; 1 = old
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y #stratify keeps proportions in young/old for train and test
)

# from sklearn.model_selection import train_test_split <-- this testing works becouse of this import
```

```python
X.isna().sum()
```

```python
X_clean = X.dropna()
y_clean = y[X_clean.index]
```

```python
X_clean.isna().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_clean, y_clean, test_size=0.3, random_state=42, stratify=y_clean
)

# from sklearn.model_selection import train_test_split <-- this testing works becouse of this import
```

```python
model = LogisticRegression() # This line works becouse of this import -> from sklearn.linear_model import LogisticRegression
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]  # P(old)

# Accuracy
acc = accuracy_score(y_test, y_pred)
# AUC
auc = roc_auc_score(y_test, y_prob)
# Confusion matrix
cm = confusion_matrix(y_test, y_pred)

print("Accuracy:", acc)
print("ROC-AUC:", auc)
print("Confusion matrix:\n", cm)


# accuracy_score, roc_auc_score, confusion_matrix works becouse of this import -> from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix
```
